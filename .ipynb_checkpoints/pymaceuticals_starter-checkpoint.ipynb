## Observations and Insights 

After processing the data a few observable trends are apparent.
Firstly, we can identify two drugs are more effective at reducing tumor size. Capomoulin and Ramicane were shown in our box plot to leave mice with the smallest volume of tumor at the final timepoint. 
Capomoulin's exact effects on mouse s185 can be found in our line graph. This graph shows a reduction of tumour size over time.
There is also a strong correlation between mouse weight and tumour size. While it is likely that a large mouse would have a large tumor, it may also be true that diet may have some effect on tumor size. 

This study's goal was to compare Capomoulin's effects against other drugs. The data seems to indicate it is a promising drug in the treatment of tumor volumen in mice. 


```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
combined_df = pd.merge(mouse_metadata, study_results, how = 'outer' , on ='Mouse ID')
# Display the data table for preview
combined_df
```

```python
# Checking the number of mice.
len(combined_df)
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 

duplicate= combined_df[combined_df.duplicated(['Mouse ID', 'Timepoint'])]
```

```python
# Optional: Get all the data for the duplicate mouse ID. 

duplicate
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
combined_df.drop(combined_df[combined_df['Mouse ID']=='g989'].index,inplace=True)
```

```python
# Checking the number of mice in the clean DataFrame.
len(combined_df)
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# calculate mean, median, std dev and SEM
tumor_volume_mean = combined_df.groupby('Drug Regimen').mean()['Tumor Volume (mm3)']
tumor_volume_median = combined_df.groupby('Drug Regimen').median()['Tumor Volume (mm3)']
tumor_volume_std = combined_df.groupby('Drug Regimen').std()['Tumor Volume (mm3)']
tumor_volume_var = combined_df.groupby('Drug Regimen').var()['Tumor Volume (mm3)']
tumor_volume_sem = combined_df.groupby('Drug Regimen').sem()['Tumor Volume (mm3)']

#create summary table
summary_stats = pd.DataFrame(tumor_volume_mean)
summary_stats['Tumor Volume Median']=tumor_volume_median
summary_stats['Tumor Volume Std Dev']=tumor_volume_std
summary_stats['Tumor Volume Variance']=tumor_volume_var
summary_stats['Tumor Volume SEM']=tumor_volume_sem

#rename first column
summary_stats = summary_stats.rename(columns={
    'Tumor Volume (mm3)': 'Tumor Volume Mean'
})


# This method produces everything in a single groupby function

summary_stats
np.round(summary_stats,decimals = 3)
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
last_timepoint = combined_df.groupby('Mouse ID').max()['Timepoint']
last_timepoint_df = pd.DataFrame(last_timepoint)
last_timepoint_df
# Merge this group df with the original dataframe to get the tumor volume at the last timepoint
time_combined_df = pd.merge(last_timepoint_df,combined_df, on = ('Mouse ID', 'Timepoint'), how = 'right')
time_combined_df.head()
```

```python
# Put treatments into a list for for loop (and later for plot labels)
# Used loc function
capomulin = time_combined_df.loc[time_combined_df["Drug Regimen"] == "Capomulin"]
ramicane = time_combined_df.loc[time_combined_df["Drug Regimen"] == "Ramicane"]
infubinol = time_combined_df.loc[time_combined_df["Drug Regimen"] == "Infubinol"]
ceftamin = time_combined_df.loc[time_combined_df["Drug Regimen"] == "Ceftamin"]


plot_Ca = capomulin["Tumor Volume (mm3)"]
plot_R = ramicane["Tumor Volume (mm3)"]
plot_I = infubinol["Tumor Volume (mm3)"]
plot_Ce = ceftamin["Tumor Volume (mm3)"]

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
#Capomulin
quartiles_Ca = plot_Ca.quantile([.25,.56,.75])
lower_Ca = quartiles_Ca[0.25]
upper_Ca = quartiles_Ca[0.75]
iqr_Ca = upper_Ca - lower_Ca
lower_bound_Ca = lower_Ca - (1.5*iqr_Ca)
upper_bound_Ca = upper_Ca + (1.5*iqr_Ca)
print('Drug: Capomoulin')
print(f"Quartiles: ")
print(quartiles_Ca)
print(f"Inter Quartile Range: {iqr_Ca}")
print(f"Anything below {lower_bound_Ca} could be outliers")
print(f"Anything above {upper_bound_Ca} could be outliers")  
print("   ")

#Ramicane
quartiles_ra = plot_R.quantile([.25,.56,.75])
lower_ra = quartiles_ra[0.25]
upper_ra = quartiles_ra[0.75]
iqr_ra = upper_ra - lower_ra
lower_bound_ra = lower_ra - (1.5*iqr_ra)
upper_bound_ra = upper_ra + (1.5*iqr_ra)
print('Drug: Ramicane')
print(f"Quartiles: ")
print(quartiles_ra)
print(f"Inter Quartile Range: {iqr_ra}")
print(f"Anything below {lower_bound_ra} could be outliers")
print(f"Anything above {upper_bound_ra} could be outliers")  
print("   ")

#Infubinol
quartiles_I = plot_I.quantile([.25,.56,.75])
lower_I = quartiles_I[0.25]
upper_I = quartiles_I[0.75]
iqr_I = upper_I - lower_I
lower_bound_I = lower_I - (1.5*iqr_I)
upper_bound_I = upper_I + (1.5*iqr_I)
print('Drug: Infubinol')
print(f"Quartiles: ")
print(quartiles_I)
print(f"Inter Quartile Range: {iqr_I}")
print(f"Anything below {lower_bound_I} could be outliers")
print(f"Anything above {upper_bound_I} could be outliers")  
print("   ")

#Ceftamin
quartiles_ce = plot_Ce.quantile([.25,.56,.75])
lower_ce = quartiles_ce[0.25]
upper_ce = quartiles_ce[0.75]
iqr_ce = upper_ce - lower_ce
lower_bound_ce = lower_ce - (1.5*iqr_ce)
upper_bound_ce = upper_ce + (1.5*iqr_ce)
print('Drug: Ceftamin')
print(f"Quartiles: ")
print(quartiles_ce)
print(f"Inter Quartile Range: {iqr_ce}")
print(f"Anything below {lower_bound_ce} could be outliers")
print(f"Anything above {upper_bound_ce} could be outliers")  

    
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest
fig1, ax1 = plt.subplots()
ax1.boxplot([plot_Ca, plot_R, plot_I, plot_Ce])
ax1.set_ylabel('Final Tumor Volume (mm3)')
ax1.set_xticklabels(["Capomulin","Ramicane", 'Infubinol','Ceftamin'])
plt.show()
plt.savefig("box plot")
```

## Line and Scatter Plots

```python
# Generate a line plot of time point versus tumor volume for a mouse treated with Capomulin
#find mice treated with Capomulin
cap_mice_df= time_combined_df.loc[time_combined_df["Drug Regimen"]== "Capomulin",:]
cap_mice_df
#first mouse on list is 's185' so we will use that as example
#get dataframe for 's185'
s185_df = time_combined_df.loc[time_combined_df["Mouse ID"]== "s185",:]
s185_df
#x_axis is timepoints
xaxis = s185_df["Timepoint"]
#yaxis is tumor size
yaxis = s185_df["Tumor Volume (mm3)"]

plt.title("Capomulin effect on mouse s185")
plt.plot(xaxis,yaxis,linewidth = 1.5, markersize = 10)
plt.xlabel("Timepoint")
plt.ylabel("Tumor Volume (mm3)")
plt.savefig('linechart')
plt.show
```

```python
# Generate a scatter plot of mouse weight versus average tumor volume for the Capomulin regimen
#use capumolin df from before
#find average tumor volume
avg_vol = cap_mice_df.groupby(['Mouse ID']).mean()
avg_vol

plt.scatter(avg_vol['Weight (g)'], avg_vol['Tumor Volume (mm3)'])
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
plt.savefig("scatter")
plt.show()
avg_vol
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
#Correlation
corr = round(st.pearsonr(avg_vol['Weight (g)'],avg_vol['Tumor Volume (mm3)'])[0],2)
print(f"The correlation between mouse weight and tumor volume is {corr}")
print("This is a strong correlation.")

#Linear regression
linereg = st.linregress(avg_vol['Weight (g)'],avg_vol['Tumor Volume (mm3)'])
linereg
```

```python
slope = 0.9544396890241049
intercept = 21.552160532685008

#line is slope * x-axis plus intercept
line = slope * (avg_vol['Weight (g)']) + intercept

plt.scatter(avg_vol['Weight (g)'], avg_vol['Tumor Volume (mm3)'])
#plot the line with formatted legend
plt.plot(avg_vol["Weight (g)"], line, 'r', label='y={:.2f}x+{:.2f}'.format(slope,intercept))
plt.xlabel('Weight(g)')
plt.xlabel ('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
plt.legend()
plt.savefig("linear regression")
plt.show()
```
